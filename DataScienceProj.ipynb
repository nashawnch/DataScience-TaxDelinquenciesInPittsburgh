# Dataset

 Our first data set, Pittsbugh Tax Delinquencies, provided data on the residential and commercial tax delinquencies in the city of Pittsburgh. We decided to focus solely on residential data in order to accurately analyze households rather than businesses. The most important data provided in this set were the exact delinquency rates for each residential address and neighborhood.

```python
import pandas as pd
```

```python
the_burgh = pd.read_csv("PittsburghTaxDelinquency.csv")
the_burgh = the_burgh[the_burgh['state_description'] == "Residential"]
the_burgh.head()
```

To make sense of our original data set, we had to seek further data to provide context to the posed questions. Our second data set, Education and Income in Pittsburgh, acquired from data.gov explores further information on the residential population of Pittsburgh including the education attainment levels, median income, and poverty rates for each neighborhood. Using this data, we aimed to draw conclusions amongst the reasons for the high tax delinquincy rates. 

```python
the_burgh2 = pd.read_excel("education--income.xls")
the_burgh2.head()
```

  By combining our two data sets, we were able to analyze tax delinquencies for each neighborhood in juxtaposition to education attainment rates, income rates, and poverty rates because we were able to analyze each neighborhood individually. 
  
  
  We first used our combined data to create two columns, called Frequency and Average Delinquency. The frequency column displayed the number of times the neighborhood was marked delinquent while the average delinquency column displays the average delinquency for each corresponding neighborhood. We then sorted our combined data in descending order, beginning with the neighborhood that was marked delinquent most frequently. 

```python
#the_burgh['Est. Percent Under Poverty (2010)'] = pd.Series(the_burgh2.groupby('Neighborhood')['Est. Percent Under Poverty (2010)'])
#the_burgh
combined = pd.merge(the_burgh, the_burgh2, left_on = 'neighborhood', right_on = 'Neighborhood')
combined.drop(['pin','_id','billing_city','council_district','ward','pli_division','police_zone',
               'fire_zone', 'pin', 'ward','tact','public_works_division',
               'Sector #','Total Pop, 25 and older (2010)', 'address'],axis=1, inplace=True)
combined['Frequency'] = combined.groupby('neighborhood')['neighborhood'].transform('count')
combined['Average Delinquency'] = combined.groupby('neighborhood')['current_delq'].transform('mean')
combined = combined.sort_values('Frequency', ascending=False)
combined
```

### Analysis
It is important to detect outliers when analyzing data and attempting to find patterns. Although each neighborhood has different delinquent rates, outliers were detected in each neighborhood. The desired_neighborhood variable holds the name of the neighborhood that you wish to find outliers in. We then filter the dataframe to omit rows that don’t correspond to that specific neighborhood. A box plot is created by plt.plot() which visually displays the delinquent outliers within that specified neighborhood. You identify the outliers by the points that are not within the region of the blue box and lines coming out of it. To find the specific residents the outliers belong to, we filter the dataframe again for the rows whose delinquent values are less than or greater than the boundary number. The return value is all rows who delinquent rates are not within the normal range for that neighborhood.

Identifying outliers within our datasets is important because they potentially skew our mean and median values which can lead to an inaccurate representation of our data. Outliers may also lead to the formation of a null hypothesis. With an null hypothesis, the perceived differences between neighborhoods could be accounted for by sampling or experimental errors such as inaccurate data in our set.
